```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# choose the data you want to load
circle_data = np.load('circle.npz')
heart_data = np.load('heart.npz')
asymmetric_data = np.load('asymmetric.npz')
```

```python
SPLIT = 0.8

# circle data
circle_X = circle_data["x"]
circle_y = circle_data["y"]
circle_X /= np.max(circle_X)  # normalize the data

# heart data
heart_X = heart_data["x"]
heart_y = heart_data["y"]
heart_X /= np.max(heart_X)  # normalize the data

# asymmetric data
asymmetric_X = asymmetric_data["x"]
asymmetric_y = asymmetric_data["y"]
asymmetric_X /= np.max(asymmetric_X)  # normalize the data
```

```python
# circle training set, validation set, & test set
circle_n_train = int(circle_X.shape[0] * SPLIT)
circle_X_train = circle_X[:circle_n_train:, :]
circle_X_valid = circle_X[circle_n_train:, :]
circle_y_train = circle_y[:circle_n_train]
circle_y_valid = circle_y[circle_n_train:]

# heart training set, validation set, & test set
heart_n_train = int(heart_X.shape[0] * SPLIT)
heart_X_train = heart_X[:heart_n_train:, :]
heart_X_valid = heart_X[heart_n_train:, :]
heart_y_train = heart_y[:heart_n_train]
heart_y_valid = heart_y[heart_n_train:]

# asymmetric training set, validation set, & test set
asymmetric_n_train = int(asymmetric_X.shape[0] * SPLIT)
asymmetric_X_train = asymmetric_X[:asymmetric_n_train:, :]
asymmetric_X_valid = asymmetric_X[asymmetric_n_train:, :]
asymmetric_y_train = asymmetric_y[:asymmetric_n_train]
asymmetric_y_valid = asymmetric_y[asymmetric_n_train:]

# circle training set, validation set, & test set using 15% data as the training dataset
circle_n_train_15 = int(circle_X.shape[0] * 0.15)
circle_X_train_15 = circle_X[:circle_n_train_15:, :]
circle_X_valid_15 = circle_X[circle_n_train_15:, :]
circle_y_train_15 = circle_y[:circle_n_train_15]
circle_y_valid_15 = circle_y[circle_n_train_15:]
```

```python
LAMBDA = 0.001
```

```python
def lstsq(A, b, lambda_=0):
    return np.linalg.solve(A.T @ A + lambda_ * np.eye(A.shape[1]), A.T @ b)
```

```python
def heatmap(f, clip=5):
    # example: heatmap(lambda x, y: x * x + y * y)
    # clip: clip the function range to [-clip, clip] to generate a clean plot
    #   set it to zero to disable this function

    xx0 = xx1 = np.linspace(np.min(X), np.max(X), 72)
    x0, x1 = np.meshgrid(xx0, xx1)
    x0, x1 = x0.ravel(), x1.ravel()
    z0 = f(x0, x1)

    if clip:
        z0[z0 > clip] = clip
        z0[z0 < -clip] = -clip

    plt.hexbin(x0, x1, C=z0, gridsize=50, cmap=cm.jet, bins=None)
    plt.colorbar()
    cs = plt.contour(
        xx0, xx1, z0.reshape(xx0.size, xx1.size), [-2, -1, -0.5, 0, 0.5, 1, 2], cmap=cm.jet)
    plt.clabel(cs, inline=1, fontsize=10)

    pos = y[:] == +1.0
    neg = y[:] == -1.0
    plt.scatter(X[pos, 0], X[pos, 1], c='red', marker='+')
    plt.scatter(X[neg, 0], X[neg, 1], c='blue', marker='v')
    plt.show()
```

```python
def assemble_feature(x, D):
    """Create a vector of polynomial features up to order D from x"""
    ### start poly_features ###
    from scipy.special import binom
    xs = []
    for d0 in range(D + 1):
        for d1 in range(D - d0 + 1):
            # non-kernel polynomial feature
            # xs.append((x[:, 0]**d0) * (x[:, 1]**d1))
            # # kernel polynomial feature
            xs.append((x[:, 0]**d0) * (x[:, 1]**d1) * np.sqrt(binom(D, d0) * binom(D - d0, d1)))
    poly_x = np.column_stack(xs)
    ### end poly_features ###
    return poly_x
```

```python
polynomials = [2, 4, 6, 8, 10, 12]
X = circle_X
y = circle_y

# circle
for D in range(2, 13):
    # training data set
    X_train = assemble_feature(circle_X_train, D)
    X_valid = assemble_feature(circle_X_valid, D)
    W = lstsq(X_train, circle_y_train, LAMBDA)
    train_err = np.mean((circle_y_train - X_train @ W)**2)  # average training squared loss
    valid_err = np.mean((circle_y_valid - X_valid @ W)**2)  # average validation squared loss

    if D in polynomials:
        heatmap(lambda x0, x1: assemble_feature(np.column_stack([x0, x1]), D) @ W)
        print("p = {:d}\nTraining Error = {:.8f}\nValidation Error = {:.8f}".format(D, train_err, valid_err))
```

```python
X_train = heart_X_train

def RBF_kernel(X, sigma):
    two_norm = np.sum(X**2, axis=1, keepdims=True) + np.sum(X_train.T**2, axis=0, keepdims=True)\
           - 2 * X @ X_train.T
    return np.exp(-two_norm/(2*(sigma**2)))
```

```python
sigmas = [10, 3, 1, 0.3, 0.1, 0.03]

for sigma in sigmas:
    K_train = RBF_kernel(heart_X_train, sigma)
    K_valid = RBF_kernel(heart_X_valid, sigma)
    W = np.linalg.solve(K_train + LAMBDA * np.eye(heart_X_train.shape[0]), heart_y_train)
    train_err = np.mean((heart_y_train - K_train @ W)**2)  # average training squared loss
    valid_err = np.mean((heart_y_valid - K_valid @ W)**2)  # average validation squared loss
    heatmap(lambda x0, x1: RBF_kernel(np.column_stack([x0, x1]), sigma) @ W)
    print("Sigma = {:.2f}\nTraining Error = {:.8f}\nValidation Error = {:.8f}".format(sigma, train_err, valid_err))
```

# 4 Eigenfaces

```python
import sklearn.datasets

dataset = sklearn.datasets.fetch_lfw_people()
X = dataset['data']
```

## (a)

```python
reshaped_X = X.reshape(13233, 62, 47)
for i in range(20):
    plt.imshow(reshaped_X[i], cmap = plt.cm.gray)
    plt.title(i+1)
    plt.show()
```

## (b)

```python
avg_X = np.average(X, axis=0)  # the average face of the dataset
center_X = X - avg_X  # center the data
plt.imshow(avg_X.reshape(62, 47), cmap = plt.cm.gray)
plt.show()
```

## (c)

```python
def PCA(n):
    cov_matrix = center_X.T @ center_X
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    top_n_eigenvalues = sorted(eigenvalues, reverse=True)[0:n]
    sorted_eigenvectors = eigenvectors[:, np.argsort(eigenvalues)[::-1]]
    top_n_eigenvectors = sorted_eigenvectors[:, 0:n]
    return top_n_eigenvectors, top_n_eigenvalues
        
def plot_eigenfaces(n, num_images=20):
    top_n_eigenvectors = PCA(n)[0]
    reduce_X = center_X @ top_n_eigenvectors
    recon_X = reduce_X @ top_n_eigenvectors.T + avg_X
    recon_X = recon_X.reshape(13233, 62, 47)
    for i in range(20):
        plt.imshow(recon_X[i], cmap = plt.cm.gray)
        plt.title(i+1)
        plt.show()    
```

```python
# plot the first 20 images reconstructed after being projected onto the top 10 principal components (PCs)
plot_eigenfaces(10, num_images=20)
```

```python
# top 100 principal components (PCs)
plot_eigenfaces(100, num_images=20)
```

```python
# top 1000 principal components (PCs)
plot_eigenfaces(1000, num_images=20)
```

```python
n = 1000
top_n_eigenvalues = PCA(n)[1]
sum_n_eigenvalues = np.sum(top_n_eigenvalues)
pc = [i+1 for i in range(len(top_n_eigenvalues))]
percentage = [np.sum(top_n_eigenvalues[0:i+1])/sum_n_eigenvalues for i in range(len(top_n_eigenvalues))]
plt.plot(pc, percentage)
plt.title('Variance Explained vs Number of PCs')
plt.xlabel('Number of PCs')
plt.ylabel('Percentage Explained')
plt.show()
```

## (f)

```python
PCs = [10, 20, 50, 100, 500, 1000, 2914]
n_train = int(X.shape[0]*0.8)
X_train = X[0:n_train, :]
X_test = X[n_train:,:]

def PCA_loss(X):
    P, D, Q = np.linalg.svd(X)
    return (np.sum(np.power(D, 2)))**(1/2)

def PCA_analysis(n, X_train, X_test):
    avg_X_train = np.average(X_train, axis = 0)
    center_X_train = X_train - avg_X_train
    avg_X_test = np.average(X_test, axis = 0)
    center_X_test = X_test - avg_X_test
    cov_matrix = center_X_train.T @ center_X_train
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    top_n_eigenvalues = sorted(eigenvalues, reverse=True)[0:n]
    sorted_eigenvectors = eigenvectors[:, np.argsort(eigenvalues)[::-1]]
    top_n_eigenvectors = sorted_eigenvectors[:, 0:n]
    reduce_X_train = center_X_train @ top_n_eigenvectors
    reduce_X_test = center_X_test @ top_n_eigenvectors
    recon_X_train = reduce_X_train @ top_n_eigenvectors.T
    recon_X_test = reduce_X_test @ top_n_eigenvectors.T
    train_loss = PCA_loss(recon_X_train - center_X_train)
    test_loss = PCA_loss(recon_X_test - center_X_test)
    return train_loss, test_loss
```

```python
train_loss_ = []
test_loss_ = []

for n in PCs:
    train_loss, test_loss = PCA_analysis(n, X_train, X_test)
    train_loss_.append(train_loss)
    test_loss_.append(test_loss)
```

```python
plt.plot(PCs, train_loss_)
plt.title('Reconstruction Loss on Training Set')
plt.xlabel('Number of PCs')
plt.ylabel('Loss')
plt.show();
```

```python
plt.plot(PCs, test_loss_)
plt.title('Reconstruction Loss on Test Set')
plt.xlabel('Number of PCs')
plt.ylabel('Loss')
plt.show();
```
